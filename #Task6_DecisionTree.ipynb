# Prediction Using Decision Tree

# Importing Libraries and loading iris dataset

```python
import numpy as np
import pandas as pd
import sklearn.datasets as datasets
import seaborn as sns
```

# Loading the dataset

```python
xyz = datasets.load_iris()
df = pd.DataFrame(xyz.data, columns=xyz.feature_names)
df.head()
```

```python
df.describe()
```

```python
df.shape
```

# Data Visualization comparing various features

```python
sns.pairplot(df)
```

```python
corr = df.corr()
sns.heatmap(corr)

```

```python
df.isnull().sum()
```

```python
df.drop(['petal width (cm)'], axis=1, inplace=True)
df.head()
```

```python
y = xyz.target
y
```

# Defining the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
tree= DecisionTreeClassifier()
tree.fit(df, y)
print("The DecisionTreeClassifier created successfully")
```

# Visualizing the Trained model


```python
#visualizing the trained decision classifier taking all 4 features in consideration
import graphviz
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(tree, out_file=dot_data, feature_names=df.columns, filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```
